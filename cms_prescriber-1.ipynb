```python
# %% [markdown]
# # Ticket 1: Project Setup & Initial Data Exploration (EDA)
#
# **Description:** Load Medicare_Part_D_Prescribers_by_Provider_and_Drug_2022-1.csv into a Jupyter notebook using Pandas. Perform initial cleaning: check/correct data types, handle missing values appropriately (especially noting suppression flags like '*' and '#'), check for duplicates. Conduct basic EDA: calculate summary statistics (df.describe()), plot distributions for key numerical columns (Tot_Drug_Cst, Tot_Clms, Tot_Day_Suply).
#
# **Goal:** Establish a clean working dataset; gain initial understanding of data structure, quality, value ranges, and distributions.
#
# **Output:** Jupyter notebook with cleaned data loading, initial descriptive statistics, and basic distribution plots (histograms, box plots).

# %% [code]
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure plotting style
sns.set_style('whitegrid')
%matplotlib inline

print("Libraries imported successfully.")

# %% [markdown]
# ## 1. Load Data
# Load the dataset from the specified CSV file. We use `low_memory=False` to prevent potential DtypeWarning due to mixed types in columns, which can happen with large files.

# %% [code]
# Define the file path
# --- IMPORTANT: Update this path to where your file is located ---
file_path = 'Medicare_Part_D_Prescribers_by_Provider_and_Drug_2022-1.csv'

try:
    # Load the dataset
    df = pd.read_csv(file_path, low_memory=False)
    print(f"Successfully loaded data from: {file_path}")
    print(f"Dataset shape: {df.shape}")
except FileNotFoundError:
    print(f"Error: The file was not found at {file_path}")
    print("Please update the 'file_path' variable with the correct location.")
    # Stop execution if file not found, or handle differently
    # For now, we'll create an empty dataframe to avoid downstream errors
    # in a real scenario, you'd likely stop execution.
    df = pd.DataFrame()
except Exception as e:
    print(f"An error occurred during loading: {e}")
    df = pd.DataFrame()

# %% [markdown]
# ## 2. Initial Inspection
# Display the first few rows, column names, data types, and check for initial missing values.

# %% [code]
if not df.empty:
    # Display the first 5 rows
    print("First 5 rows of the dataset:")
    display(df.head())

    # Display basic information (column names, non-null counts, data types)
    print("\nDataset Info:")
    df.info()

    # Check initial missing values (before type conversion)
    print("\nInitial Missing Values per Column:")
    initial_missing = df.isnull().sum()
    print(initial_missing[initial_missing > 0])
else:
    print("DataFrame is empty. Skipping initial inspection.")

# %% [markdown]
# ## 3. Data Cleaning
#
# ### 3.1 Correct Data Types
# Identify key numerical columns (`Tot_Clms`, `Tot_Day_Suply`, `Tot_Drug_Cst`) which might be loaded as object type due to suppression flags ('*', '#') or other non-numeric characters. Convert them to numeric, coercing errors into NaN. This effectively handles the suppression flags by treating them as missing/unreliable numeric data.

# %% [code]
if not df.empty:
    # Define columns to convert to numeric
    # Note: Verify these column names match your CSV *exactly*
    numeric_cols = ['Tot_Clms', 'Tot_Day_Suply', 'Tot_Drug_Cst']
    
    # Check if columns exist before attempting conversion
    cols_to_convert = [col for col in numeric_cols if col in df.columns]
    missing_cols = [col for col in numeric_cols if col not in df.columns]
    
    if missing_cols:
         print(f"\nWarning: The following expected numeric columns were not found: {missing_cols}")

    if cols_to_convert:
        print(f"\nConverting columns to numeric: {cols_to_convert}")
        for col in cols_to_convert:
            # Check current dtype
            print(f"  Converting '{col}' (current dtype: {df[col].dtype})...")
            # Use pd.to_numeric, coercing errors to NaN
            df[col] = pd.to_numeric(df[col], errors='coerce')
            # Verify new dtype
            print(f"  '{col}' new dtype: {df[col].dtype}")
        
        print("\nData types after conversion:")
        df.info() # Display info again to see changed types
    else:
        print("\nNo columns found to convert to numeric.")
else:
    print("DataFrame is empty. Skipping data type correction.")


# %% [markdown]
# ### 3.2 Handle Missing Values (Post-Conversion)
# Check for missing values again after numeric conversion. The `NaN` values in the converted columns now represent both originally missing data and the suppressed values ('*', '#'). We will note the count of these NaNs.

# %% [code]
if not df.empty and cols_to_convert:
    print("\nMissing Values per Column (after numeric conversion):")
    missing_after = df.isnull().sum()
    missing_counts = missing_after[missing_after > 0]

    if not missing_counts.empty:
        print(missing_counts)
        print("\nNote: NaN values in numeric columns may represent originally missing data or suppressed values ('*', '#') from the source file.")
    else:
        print("No missing values found after conversion.")
elif df.empty:
     print("DataFrame is empty. Skipping missing value check.")
else: # Dataframe not empty, but no columns were converted
    print("\nMissing Values Check (no numeric conversion performed):")
    initial_missing = df.isnull().sum()
    missing_counts = initial_missing[initial_missing > 0]
    if not missing_counts.empty:
        print(missing_counts)
    else:
        print("No missing values found.")

# %% [markdown]
# ### 3.3 Check for Duplicate Rows
# Identify the number of complete duplicate rows in the dataset. Depending on the analysis goals, these might be removed later.

# %% [code]
if not df.empty:
    num_duplicates = df.duplicated().sum()
    print(f"\nNumber of complete duplicate rows found: {num_duplicates}")
    if num_duplicates > 0:
        print("Consider investigating or removing these duplicates in subsequent steps if appropriate.")
else:
    print("DataFrame is empty. Skipping duplicate check.")

# %% [markdown]
# ## 4. Basic Exploratory Data Analysis (EDA)
#
# ### 4.1 Summary Statistics
# Calculate descriptive statistics for the key numerical columns.

# %% [code]
if not df.empty and cols_to_convert:
    print("\nSummary Statistics for Key Numerical Columns:")
    # Use the list of columns that were successfully converted
    display(df[cols_to_convert].describe())
elif not df.empty:
     print("\nSkipping numerical summary statistics as key columns were not found or converted.")
else:
    print("DataFrame is empty. Skipping summary statistics.")

# %% [code]
# Also show summary statistics for object/categorical columns (optional but useful)
if not df.empty:
    print("\nSummary Statistics for Object/Categorical Columns:")
    try:
        display(df.describe(include='object'))
    except Exception as e:
        print(f"Could not generate object statistics: {e}")
else:
     print("DataFrame is empty. Skipping object summary statistics.")

# %% [markdown]
# ### 4.2 Plot Distributions
# Visualize the distributions of the key numerical columns using histograms and box plots. This helps understand the spread, central tendency, and identify potential outliers or skewness.

# %% [code]
if not df.empty and cols_to_convert:
    print("\nPlotting Distributions for Key Numerical Columns...")

    for col in cols_to_convert:
        plt.figure(figsize=(12, 5))

        # Histogram
        plt.subplot(1, 2, 1)
        sns.histplot(df[col], kde=True, bins=50) # Added kde=True for smooth curve
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')

        # Box Plot
        plt.subplot(1, 2, 2)
        sns.boxplot(x=df[col])
        plt.title(f'Box Plot of {col}')
        plt.xlabel(col)

        plt.tight_layout()
        plt.show()

        # Check for skewness (often high for cost/count data)
        if df[col].notna().any(): # Check if there are any non-NaN values before calculating skew
             skewness = df[col].skew()
             print(f"Skewness for {col}: {skewness:.2f}")
             if abs(skewness) > 1:
                  print(f"  Note: Distribution for {col} is highly skewed. Consider log transformation for visualization/modeling if needed.")
        else:
             print(f"  Cannot calculate skewness for {col} as it contains only NaN values.")


elif df.empty:
    print("DataFrame is empty. Skipping plotting.")
else:
    print("Skipping plotting as key numerical columns were not found or converted.")


# %% [markdown]
# ## 5. Summary & Next Steps
#
# - The dataset has been loaded.
# - Initial inspection performed (shape, info, head).
# - Key numerical columns (`Tot_Clms`, `Tot_Day_Suply`, `Tot_Drug_Cst`) converted to numeric type, handling suppression flags ('*', '#') by converting them to NaN.
# - Missing values (including suppressed data) have been identified and quantified.
# - Duplicate rows have been counted.
# - Basic descriptive statistics calculated.
# - Distributions of key numerical columns visualized using histograms and box plots. Potential skewness noted.
#
# **Next Steps:** Proceed to further data cleaning, feature engineering, or more in-depth analysis as required by subsequent tickets. The cleaned data (potentially with NaNs still present, representing suppression/missingness) is ready for the next phase.

# %% [code]
# Optional: Save the cleaned dataframe if needed for the next step
# if not df.empty:
#     cleaned_file_path = 'medicare_part_d_2022_cleaned_step1.csv'
#     # df.to_csv(cleaned_file_path, index=False)
#     # print(f"\nCleaned dataframe (intermediate step) could be saved to: {cleaned_file_path}")
#     pass # Uncomment above lines to save

print("\nTicket 1 Execution Complete.")
```

```python
# %% [markdown]
# # Ticket 2: Identify High-Cost & High-Volume Drugs
#
# **Description:** Group data by Gnrc_Name and Brnd_Name. Calculate total Tot_Drug_Cst and total Tot_Clms for each drug. Rank drugs to identify the top contributors to overall cost and volume. Calculate Cost_Per_Claim (Tot_Drug_Cst / Tot_Clms) for each drug/provider combination and summarize per drug (e.g., average Cost_Per_Claim).
#
# **Goal:** Pinpoint the specific drugs driving the most significant spend and utilization, which are primary targets for negotiation focus. Understand cost efficiency per claim.
#
# **Output:** Tables/DataFrames in Jupyter showing ranked lists of drugs by total cost, total claims, and average cost per claim.

# %% [markdown]
# ## Prerequisite Check
# Ensure the DataFrame `df` from Ticket 1 exists and contains the necessary columns.

# %% [code]
# Check if df exists and has required columns
required_cols = ['Gnrc_Name', 'Brnd_Name', 'Tot_Drug_Cst', 'Tot_Clms']
df_exists = 'df' in locals() or 'df' in globals()

if df_exists and isinstance(df, pd.DataFrame) and all(col in df.columns for col in required_cols):
    print("DataFrame 'df' found with required columns.")
    # Optionally display info again if needed
    # df[required_cols].info()
elif not df_exists or not isinstance(df, pd.DataFrame):
    print("Error: DataFrame 'df' not found or is not a Pandas DataFrame.")
    print("Please ensure Ticket 1 has been executed successfully.")
    # Create an empty df to prevent further errors, or stop execution
    df = pd.DataFrame(columns=required_cols) # Allow code below to run without erroring, but produce empty results
else: # df exists but missing columns
    missing_req_cols = [col for col in required_cols if col not in df.columns]
    print(f"Error: DataFrame 'df' is missing required columns: {missing_req_cols}")
    print("Please ensure Ticket 1 processed the data correctly.")
    # Keep df as is, but subsequent steps might fail or give warnings

# %% [markdown]
# ## 1. Aggregate by Generic Name (Gnrc_Name)
# Group the data by generic drug name and calculate the total cost and total claims for each.

# %% [code]
if not df.empty and all(col in df.columns for col in ['Gnrc_Name', 'Tot_Drug_Cst', 'Tot_Clms']):
    print("Aggregating data by Generic Name (Gnrc_Name)...")
    # Group by Generic Name and sum up Costs and Claims
    # Note: sum() skips NaN values by default
    generic_agg = df.groupby('Gnrc_Name', dropna=False).agg( # Keep NaN Gnrc_Name if exists
        Total_Cost=('Tot_Drug_Cst', 'sum'),
        Total_Claims=('Tot_Clms', 'sum')
    ).reset_index() # Convert grouped output back to DataFrame

    print(f"Aggregation complete. Found {generic_agg.shape[0]} unique generic names.")
else:
    print("Skipping Generic Name aggregation due to missing DataFrame or columns.")
    generic_agg = pd.DataFrame(columns=['Gnrc_Name', 'Total_Cost', 'Total_Claims'])


# %% [markdown]
# ## 2. Rank Generic Drugs
# Rank the aggregated generic drugs by total cost and total claims.

# %% [code]
if not generic_agg.empty:
    # Rank by Total Cost (descending)
    generic_ranked_cost = generic_agg.sort_values(by='Total_Cost', ascending=False)

    # Rank by Total Claims (descending)
    generic_ranked_claims = generic_agg.sort_values(by='Total_Claims', ascending=False)

    # Display top N (e.g., Top 20)
    N = 20
    print(f"\nTop {N} Generic Drugs by Total Cost:")
    display(generic_ranked_cost.head(N))

    print(f"\nTop {N} Generic Drugs by Total Claims:")
    display(generic_ranked_claims.head(N))
else:
    print("Skipping Generic Name ranking as aggregation results are empty.")

# %% [markdown]
# ## 3. Aggregate by Brand Name (Brnd_Name)
# Group the data by brand drug name and calculate the total cost and total claims for each.

# %% [code]
if not df.empty and all(col in df.columns for col in ['Brnd_Name', 'Tot_Drug_Cst', 'Tot_Clms']):
    print("Aggregating data by Brand Name (Brnd_Name)...")
    # Group by Brand Name and sum up Costs and Claims
    brand_agg = df.groupby('Brnd_Name', dropna=False).agg( # Keep NaN Brnd_Name if exists
        Total_Cost=('Tot_Drug_Cst', 'sum'),
        Total_Claims=('Tot_Clms', 'sum')
    ).reset_index()

    print(f"Aggregation complete. Found {brand_agg.shape[0]} unique brand names.")
else:
    print("Skipping Brand Name aggregation due to missing DataFrame or columns.")
    brand_agg = pd.DataFrame(columns=['Brnd_Name', 'Total_Cost', 'Total_Claims'])

# %% [markdown]
# ## 4. Rank Brand Drugs
# Rank the aggregated brand drugs by total cost and total claims.

# %% [code]
if not brand_agg.empty:
    # Rank by Total Cost (descending)
    brand_ranked_cost = brand_agg.sort_values(by='Total_Cost', ascending=False)

    # Rank by Total Claims (descending)
    brand_ranked_claims = brand_agg.sort_values(by='Total_Claims', ascending=False)

    # Display top N (e.g., Top 20)
    N = 20
    print(f"\nTop {N} Brand Drugs by Total Cost:")
    display(brand_ranked_cost.head(N))

    print(f"\nTop {N} Brand Drugs by Total Claims:")
    display(brand_ranked_claims.head(N))
else:
    print("Skipping Brand Name ranking as aggregation results are empty.")


# %% [markdown]
# ## 5. Calculate Cost Per Claim
# Calculate the cost per claim for each individual prescriber/drug row in the original DataFrame.
# Handle potential division by zero or NaN values.

# %% [code]
if not df.empty and all(col in df.columns for col in ['Tot_Drug_Cst', 'Tot_Clms']):
    print("\nCalculating Cost Per Claim for each row...")
    # Use np.divide for slightly better control over division by zero/NaN
    # np.divide handles 0 denominator by returning inf, which is mathematically sound here.
    # It also handles NaN correctly (NaN numerator or denominator -> NaN result)
    df['Cost_Per_Claim'] = np.divide(df['Tot_Drug_Cst'], df['Tot_Clms'])

    # Optional: Replace infinite values with NaN if preferred
    # df['Cost_Per_Claim'].replace([np.inf, -np.inf], np.nan, inplace=True)
    # print("Replaced infinite Cost_Per_Claim values with NaN.") # Uncomment if using replace

    print("Cost_Per_Claim column added/updated.")
    # Display some rows with the new column
    display(df[['Gnrc_Name', 'Brnd_Name', 'Tot_Drug_Cst', 'Tot_Clms', 'Cost_Per_Claim']].head())

    # Check how many NaNs/Infs are in the new column
    cpc_nan_count = df['Cost_Per_Claim'].isnull().sum()
    cpc_inf_count = np.isinf(df['Cost_Per_Claim']).sum()
    print(f"Cost_Per_Claim column contains {cpc_nan_count} NaN values.")
    print(f"Cost_Per_Claim column contains {cpc_inf_count} infinite values (due to 0 claims).")

else:
    print("Skipping Cost Per Claim calculation due to missing DataFrame or columns.")
    # Ensure the column exists even if empty for subsequent steps
    if 'df' in locals() or 'df' in globals() and isinstance(df, pd.DataFrame):
         df['Cost_Per_Claim'] = np.nan


# %% [markdown]
# ## 6. Summarize Average Cost Per Claim by Drug
# Calculate the average cost per claim for each generic and brand drug by averaging the `Cost_Per_Claim` values calculated above across all providers for that drug.

# %% [code]
if not df.empty and 'Cost_Per_Claim' in df.columns:
    print("\nCalculating Average Cost Per Claim per Generic Name...")
    # Group by Generic Name and calculate the mean of Cost_Per_Claim
    # Note: mean() skips NaN values by default
    generic_avg_cpc = df.groupby('Gnrc_Name', dropna=False)['Cost_Per_Claim'].mean().reset_index()
    generic_avg_cpc.rename(columns={'Cost_Per_Claim': 'Average_Cost_Per_Claim'}, inplace=True)
    generic_avg_cpc = generic_avg_cpc.sort_values(by='Average_Cost_Per_Claim', ascending=False)

    print(f"\nTop {N} Generic Drugs by Average Cost Per Claim:")
    display(generic_avg_cpc.head(N))

    print("\nCalculating Average Cost Per Claim per Brand Name...")
    # Group by Brand Name and calculate the mean of Cost_Per_Claim
    brand_avg_cpc = df.groupby('Brnd_Name', dropna=False)['Cost_Per_Claim'].mean().reset_index()
    brand_avg_cpc.rename(columns={'Cost_Per_Claim': 'Average_Cost_Per_Claim'}, inplace=True)
    brand_avg_cpc = brand_avg_cpc.sort_values(by='Average_Cost_Per_Claim', ascending=False)

    print(f"\nTop {N} Brand Drugs by Average Cost Per Claim:")
    display(brand_avg_cpc.head(N))

    # Alternative Calculation: Total Cost / Total Claims per drug
    # This gives a potentially different, weighted average cost per claim.
    if not generic_agg.empty:
         generic_agg['Weighted_Avg_Cost_Per_Claim'] = np.divide(generic_agg['Total_Cost'], generic_agg['Total_Claims'])
         print(f"\nTop {N} Generic Drugs by Weighted Average Cost Per Claim (Total Cost / Total Claims):")
         display(generic_agg.sort_values(by='Weighted_Avg_Cost_Per_Claim', ascending=False).head(N)[['Gnrc_Name', 'Weighted_Avg_Cost_Per_Claim']])

    if not brand_agg.empty:
         brand_agg['Weighted_Avg_Cost_Per_Claim'] = np.divide(brand_agg['Total_Cost'], brand_agg['Total_Claims'])
         print(f"\nTop {N} Brand Drugs by Weighted Average Cost Per Claim (Total Cost / Total Claims):")
         display(brand_agg.sort_values(by='Weighted_Avg_Cost_Per_Claim', ascending=False).head(N)[['Brnd_Name', 'Weighted_Avg_Cost_Per_Claim']])

else:
    print("Skipping Average Cost Per Claim summary due to missing DataFrame or 'Cost_Per_Claim' column.")


# %% [markdown]
# ## 7. Summary & Next Steps
#
# - Aggregated total cost and claims by both Generic Name and Brand Name.
# - Ranked drugs based on these aggregate metrics to identify top contributors.
# - Calculated `Cost_Per_Claim` for each provider-drug record.
# - Calculated the average `Cost_Per_Claim` across providers for each Generic and Brand Name.
# - Additionally calculated a 'Weighted Average Cost Per Claim' (Total Cost / Total Claims) for comparison.
#
# **Next Steps:** These ranked lists and cost-per-claim metrics provide key insights for identifying negotiation targets. This data can be further analyzed (e.g., geographical variations, prescriber specialty analysis) or prepared for visualization in Tableau.

# %% [code]
# Optional: You might want to merge some of these results or save them
# Example: Merge average CPC back to the main aggregated tables
# if not generic_agg.empty and not generic_avg_cpc.empty:
#     generic_summary = pd.merge(generic_ranked_cost, generic_avg_cpc, on='Gnrc_Name', how='left')
#     generic_summary = pd.merge(generic_summary, generic_agg[['Gnrc_Name', 'Weighted_Avg_Cost_Per_Claim']], on='Gnrc_Name', how='left')
#     print("\nCombined Generic Drug Summary Table (Top 20 by Cost):")
#     display(generic_summary.head(20))
    # generic_summary.to_csv('medicare_part_d_2022_generic_summary.csv', index=False)

# if not brand_agg.empty and not brand_avg_cpc.empty:
#     brand_summary = pd.merge(brand_ranked_cost, brand_avg_cpc, on='Brnd_Name', how='left')
#     brand_summary = pd.merge(brand_summary, brand_agg[['Brnd_Name', 'Weighted_Avg_Cost_Per_Claim']], on='Brnd_Name', how='left')
#     print("\nCombined Brand Drug Summary Table (Top 20 by Cost):")
#     display(brand_summary.head(20))
    # brand_summary.to_csv('medicare_part_d_2022_brand_summary.csv', index=False)

print("\nTicket 2 Execution Complete.")
```

```python
# %% [markdown]
# # Ticket 3: Analyze Brand vs. Generic Cost & Utilization
#
# **Description:** Identify brand drugs with corresponding generic entries in the dataset. For these pairs, compare total Tot_Drug_Cst, total Tot_Clms, and average Cost_Per_Claim. Calculate generic dispensing rates (GDR) based on claim volume (Generic Claims / (Generic Claims + Brand Claims)) and potentially cost impact.
#
# **Goal:** Quantify the cost differences and usage patterns between brand drugs and their generic alternatives, highlighting potential savings and negotiation leverage.
#
# **Output:** Comparative tables/plots in Jupyter for brand/generic pairs. Calculated GDR metrics.

# %% [markdown]
# ## Prerequisite Check
# Ensure the DataFrame `df` exists and contains the necessary columns, including `Cost_Per_Claim` from Ticket 2.

# %% [code]
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Check if df exists and has required columns
required_cols_t3 = ['Gnrc_Name', 'Brnd_Name', 'Tot_Drug_Cst', 'Tot_Clms', 'Cost_Per_Claim']
df_exists_t3 = 'df' in locals() or 'df' in globals()

if df_exists_t3 and isinstance(df, pd.DataFrame) and all(col in df.columns for col in required_cols_t3):
    print("DataFrame 'df' found with required columns for Ticket 3.")
    # Make a copy to avoid modifying the original df unintentionally if needed later
    df_t3 = df.copy()
elif not df_exists_t3 or not isinstance(df, pd.DataFrame):
    print("Error: DataFrame 'df' not found or is not a Pandas DataFrame.")
    print("Please ensure Tickets 1 and 2 have been executed successfully.")
    df_t3 = pd.DataFrame(columns=required_cols_t3) # Allow code below to run without erroring
else: # df exists but missing columns
    missing_req_cols_t3 = [col for col in required_cols_t3 if col not in df.columns]
    print(f"Error: DataFrame 'df' is missing required columns: {missing_req_cols_t3}")
    print("Please ensure Tickets 1 and 2 processed the data correctly.")
    # Attempt to continue with available columns, subsequent steps might fail
    df_t3 = df.copy()


# %% [markdown]
# ## 1. Identify Generic Names with Both Brand and Generic Entries
# We identify `Gnrc_Name` values that are associated with more than one unique `Brnd_Name`. We assume the entry where `Brnd_Name` matches `Gnrc_Name` is the generic form.

# %% [code]
if not df_t3.empty and 'Gnrc_Name' in df_t3.columns and 'Brnd_Name' in df_t3.columns:
    # Drop rows where Gnrc_Name is NaN as they cannot form pairs
    df_t3_cleaned = df_t3.dropna(subset=['Gnrc_Name'])

    # Group by Generic Name and count unique Brand Names
    brand_counts = df_t3_cleaned.groupby('Gnrc_Name')['Brnd_Name'].nunique()

    # Filter for Generic Names with more than one associated Brand Name
    multi_brand_generics = brand_counts[brand_counts > 1].index.tolist()

    print(f"Found {len(multi_brand_generics)} generic names associated with multiple brand names (potential Brand/Generic pairs).")

    if len(multi_brand_generics) < 20: # Print some examples if the list is small
         print("Examples:", multi_brand_generics[:5])
else:
    print("Skipping identification of pairs due to missing DataFrame or columns.")
    multi_brand_generics = []


# %% [markdown]
# ## 2. Filter Data for Potential Pairs
# Create a DataFrame containing only the rows corresponding to the identified generic names.

# %% [code]
if multi_brand_generics:
    # Filter the original cleaned dataframe (df_t3_cleaned)
    df_pairs = df_t3_cleaned[df_t3_cleaned['Gnrc_Name'].isin(multi_brand_generics)].copy() # Use copy to avoid SettingWithCopyWarning
    print(f"Filtered DataFrame 'df_pairs' created with {df_pairs.shape[0]} rows for analysis.")
else:
    print("No potential Brand/Generic pairs identified. Skipping further analysis.")
    df_pairs = pd.DataFrame(columns=required_cols_t3) # Create empty df

# %% [markdown]
# ## 3. Aggregate Metrics and Label Brand/Generic
# Group the filtered data by `Gnrc_Name` and `Brnd_Name` to get summary statistics. Label each row as 'Brand' or 'Generic' based on our assumption (`Brnd_Name == Gnrc_Name`).

# %% [code]
if not df_pairs.empty:
    print("Aggregating metrics for Brand/Generic pairs...")
    pairs_agg = df_pairs.groupby(['Gnrc_Name', 'Brnd_Name']).agg(
        Total_Cost=('Tot_Drug_Cst', 'sum'),
        Total_Claims=('Tot_Clms', 'sum'),
        # Calculate mean of Cost_Per_Claim across providers for this drug
        # Note: This is the average of ratios, not the ratio of sums
        Average_CPC=('Cost_Per_Claim', 'mean')
    ).reset_index()

    # Label Drug Type
    # Ensure comparison handles potential dtype differences or spaces if any
    pairs_agg['Drug_Type'] = np.where(
        pairs_agg['Gnrc_Name'].str.strip().str.lower() == pairs_agg['Brnd_Name'].str.strip().str.lower(),
        'Generic',
        'Brand'
    )

    # Alternative: Calculate Weighted Average CPC (Total Cost / Total Claims)
    pairs_agg['Weighted_Avg_CPC'] = np.divide(pairs_agg['Total_Cost'], pairs_agg['Total_Claims'])
    # Replace inf with NaN if preferred
    pairs_agg['Weighted_Avg_CPC'].replace([np.inf, -np.inf], np.nan, inplace=True)


    print("Aggregation and labeling complete.")
    display(pairs_agg.head())

    # Check counts of Brand vs Generic labels
    print("\nCounts by Drug Type:")
    print(pairs_agg['Drug_Type'].value_counts())

    # Check if any Gnrc_Name ended up with only 'Brand' labels (edge case)
    check_labels = pairs_agg.groupby('Gnrc_Name')['Drug_Type'].apply(lambda x: set(x))
    only_brands = check_labels[check_labels.apply(lambda s: s == {'Brand'})]
    if not only_brands.empty:
        print(f"\nWarning: {len(only_brands)} Gnrc_Names have multiple Brnd_Names but none matched Gnrc_Name exactly.")
        print("These might need manual review or a different labeling logic.")
        # print(only_brands.head()) # Uncomment to see examples

else:
    print("Skipping aggregation and labeling as df_pairs is empty.")
    pairs_agg = pd.DataFrame()


# %% [markdown]
# ## 4. Comparative Table
# Display the aggregated data sorted to easily compare Brand and Generic versions side-by-side.

# %% [code]
if not pairs_agg.empty:
    # Sort for easy comparison
    pairs_sorted = pairs_agg.sort_values(by=['Gnrc_Name', 'Drug_Type'], ascending=[True, False]) # Show Brand then Generic

    print("\nComparative Table (Brand vs. Generic - Sample):")
    # Display a sample focusing on a few generic names to show the structure
    sample_generics = pairs_sorted['Gnrc_Name'].unique()[:5] # Show first 5 Gnrc_Names
    display(pairs_sorted[pairs_sorted['Gnrc_Name'].isin(sample_generics)])

    # Optionally display the full table or save it
    # display(pairs_sorted)
    # pairs_sorted.to_csv('brand_generic_comparison.csv', index=False)
else:
    print("Skipping comparative table display.")


# %% [markdown]
# ## 5. Calculate Generic Dispensing Rate (GDR)
# Calculate GDR based on claim volume and cost for each `Gnrc_Name` that has both Brand and Generic entries identified.

# %% [code]
if not pairs_agg.empty:
    print("\nCalculating Generic Dispensing Rates (GDR)...")

    # Aggregate claims and cost by Gnrc_Name and Drug_Type
    gdr_data = pairs_agg.groupby(['Gnrc_Name', 'Drug_Type']).agg(
        Group_Total_Claims=('Total_Claims', 'sum'),
        Group_Total_Cost=('Total_Cost', 'sum')
    ).reset_index()

    # Pivot the table to get Brand and Generic metrics side-by-side for calculation
    gdr_pivot = gdr_data.pivot(index='Gnrc_Name', columns='Drug_Type', values=['Group_Total_Claims', 'Group_Total_Cost']).fillna(0)

    # Flatten MultiIndex columns (e.g., ('Group_Total_Claims', 'Brand') -> 'Claims_Brand')
    gdr_pivot.columns = ['_'.join(col).strip() for col in gdr_pivot.columns.values]
    gdr_pivot.reset_index(inplace=True)

    # Ensure all expected columns exist after pivot, adding them with 0 if missing
    expected_gdr_cols = ['Group_Total_Claims_Brand', 'Group_Total_Claims_Generic',
                         'Group_Total_Cost_Brand', 'Group_Total_Cost_Generic']
    for col in expected_gdr_cols:
        if col not in gdr_pivot.columns:
            gdr_pivot[col] = 0
            print(f"  Note: Added missing column '{col}' with zeros (likely means no Brand/Generic found for some groups after filtering).")


    # Calculate Total Claims and Total Cost per Generic Name
    gdr_pivot['Total_Claims_Pair'] = gdr_pivot['Group_Total_Claims_Brand'] + gdr_pivot['Group_Total_Claims_Generic']
    gdr_pivot['Total_Cost_Pair'] = gdr_pivot['Group_Total_Cost_Brand'] + gdr_pivot['Group_Total_Cost_Generic']

    # Calculate GDR based on Claims Volume
    # Use np.divide to handle potential division by zero (returns NaN)
    gdr_pivot['GDR_Claims'] = np.divide(
        gdr_pivot['Group_Total_Claims_Generic'],
        gdr_pivot['Total_Claims_Pair']
    ) * 100 # Express as percentage

    # Calculate GDR based on Cost
    gdr_pivot['GDR_Cost'] = np.divide(
        gdr_pivot['Group_Total_Cost_Generic'],
        gdr_pivot['Total_Cost_Pair']
    ) * 100 # Express as percentage

    # Select and display relevant columns
    gdr_results = gdr_pivot[[
        'Gnrc_Name',
        'Group_Total_Claims_Generic', 'Group_Total_Claims_Brand', 'Total_Claims_Pair', 'GDR_Claims',
        'Group_Total_Cost_Generic', 'Group_Total_Cost_Brand', 'Total_Cost_Pair', 'GDR_Cost'
    ]].sort_values(by='Total_Cost_Pair', ascending=False) # Sort by total cost of the pair

    print("\nGeneric Dispensing Rate (GDR) Results (Top Pairs by Total Cost):")
    display(gdr_results.head(20)) # Display top 20

    # Optionally save GDR results
    # gdr_results.to_csv('gdr_analysis_results.csv', index=False)

else:
    print("Skipping GDR calculation as aggregated pairs data is empty.")
    gdr_results = pd.DataFrame()


# %% [markdown]
# ## 6. Visualization (Optional Example)
# Create a simple plot comparing Brand vs Generic average cost per claim for a few high-cost pairs.

# %% [code]
if not pairs_agg.empty and not gdr_results.empty:
    print("\nVisualizing Brand vs Generic Average CPC for Top 10 Costliest Pairs...")

    # Get the Gnrc_Names of the top 10 costliest pairs from gdr_results
    top_10_cost_pairs_generics = gdr_results.nlargest(10, 'Total_Cost_Pair')['Gnrc_Name'].tolist()

    # Filter the `pairs_agg` table for these generics
    plot_data = pairs_agg[pairs_agg['Gnrc_Name'].isin(top_10_cost_pairs_generics)]

    # Create the plot
    plt.figure(figsize=(12, 7))
    sns.barplot(data=plot_data, x='Average_CPC', y='Gnrc_Name', hue='Drug_Type', orient='h',
                order=top_10_cost_pairs_generics, # Keep the order from costliest pair
                palette={'Brand': 'salmon', 'Generic': 'lightgreen'}) # Use distinct colors

    plt.title('Brand vs. Generic Average Cost Per Claim (Top 10 Pairs by Total Spend)')
    plt.xlabel('Average Cost Per Claim ($)')
    plt.ylabel('Generic Name')
    plt.legend(title='Drug Type')
    plt.tight_layout()
    plt.show()

    # Another plot: GDR Claims vs GDR Cost
    plt.figure(figsize=(8, 8))
    sns.scatterplot(data=gdr_results.fillna(0), x='GDR_Claims', y='GDR_Cost',
                    size='Total_Cost_Pair',  # Size dots by total cost of the pair
                    hue=gdr_results['Total_Cost_Pair'].apply(np.log10), # Color dots by cost magnitude
                    sizes=(20, 1000), alpha=0.6)
    plt.title('Generic Dispensing Rate (GDR) by Claims vs. Cost')
    plt.xlabel('GDR by Claims Volume (%)')
    plt.ylabel('GDR by Cost (%)')
    plt.plot([0, 100], [0, 100], 'k--', alpha=0.5) # Add diagonal line for reference
    plt.xlim(-5, 105)
    plt.ylim(-5, 105)
    plt.legend('',frameon=False) # Hide legend auto-generated by size/hue for cleaner look
    plt.grid(True)
    plt.show()
    print("Note: In the scatter plot, dot size represents the total cost of the Brand/Generic pair.")

else:
    print("Skipping visualization as data is unavailable.")


# %% [markdown]
# ## 7. Summary & Next Steps
#
# - Identified generic drugs (`Gnrc_Name`) that have both brand and generic (`Brnd_Name`) entries in the dataset.
# - Aggregated cost, claims, and average cost-per-claim metrics for these Brand/Generic pairs.
# - Created comparative tables showing these metrics side-by-side.
# - Calculated Generic Dispensing Rates (GDR) based on both claims volume and total cost.
# - Visualized the difference in average cost-per-claim and the distribution of GDRs.
#
# **Next Steps:** This analysis directly highlights opportunities where shifting utilization from Brand to Generic could yield savings. The GDR metrics quantify current adoption levels. This information is valuable for formulary discussions, provider education, and rebate negotiations (e.g., negotiating higher rebates on brands with low GDR or where generic alternatives are significantly cheaper). The output tables (`pairs_sorted`, `gdr_results`) can be exported for further analysis or use in Tableau.

# %% [code]
print("\nTicket 3 Execution Complete.")
```

```python
# %% [markdown]
# # Ticket 5: Detect Cost & Volume Outliers
#
# **Description:** For key drugs or within specific specialties, analyze the distribution of Cost_Per_Claim and Tot_Clms per provider. Use statistical methods (e.g., calculating IQR or Z-scores) or visualizations (box plots) to identify providers with significantly higher cost per claim or prescription volume compared to peers.
#
# **Goal:** Flag potential anomalies in prescribing cost or volume that might warrant further investigation or specific discussion points in negotiations.
#
# **Output:** Lists or highlighted visualizations in Jupyter showing potential outlier providers/prescriptions based on defined criteria.

# %% [markdown]
# ## Prerequisite Check
# Ensure the DataFrame `df` exists and contains necessary columns. Check specifically for a provider specialty column (e.g., `Prscrbr_Type`).

# %% [code]
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Check if df exists and has required columns
# --- ADJUST 'Prscrbr_Type' if your specialty column name is different ---
specialty_col = 'Prscrbr_Type'
required_cols_t5 = ['Prscrbr_NPI', 'Gnrc_Name', 'Brnd_Name', 'Tot_Clms', 'Cost_Per_Claim', specialty_col]
df_exists_t5 = 'df' in locals() or 'df' in globals()
specialty_col_exists = False

if df_exists_t5 and isinstance(df, pd.DataFrame):
    # Check for essential columns first
    essential_cols = ['Prscrbr_NPI', 'Gnrc_Name', 'Brnd_Name', 'Tot_Clms', 'Cost_Per_Claim']
    if all(col in df.columns for col in essential_cols):
        print("DataFrame 'df' found with essential columns for Ticket 5.")
        df_t5 = df.copy()
        # Check for specialty column
        if specialty_col in df.columns:
             specialty_col_exists = True
             print(f"Specialty column '{specialty_col}' found.")
             # Optional: Check number of unique specialties
             unique_specialties = df_t5[specialty_col].nunique()
             print(f"  Found {unique_specialties} unique specialties.")
             if unique_specialties > 100: # Example threshold
                 print(f"  Warning: High number of unique specialties. Consider cleaning or grouping '{specialty_col}' for effective analysis.")
             # Handle missing specialties if necessary
             df_t5[specialty_col].fillna('Unknown', inplace=True)

        else:
             print(f"Warning: Specialty column '{specialty_col}' not found. Outlier analysis will not be grouped by specialty.")
             # Add a placeholder if needed downstream, or adjust logic
             # df_t5[specialty_col] = 'N/A' # Example placeholder
    else:
        missing_ess_cols = [col for col in essential_cols if col not in df.columns]
        print(f"Error: DataFrame 'df' is missing essential columns: {missing_ess_cols}")
        df_t5 = pd.DataFrame(columns=essential_cols + ([specialty_col] if specialty_col_exists else [])) # Empty DF
elif not df_exists_t5 or not isinstance(df, pd.DataFrame):
    print("Error: DataFrame 'df' not found or is not a Pandas DataFrame.")
    print("Please ensure Tickets 1-4 have been executed successfully.")
    df_t5 = pd.DataFrame(columns=essential_cols + [specialty_col]) # Empty DF
else: # Should not be reached based on above logic, but for completeness
     df_t5 = pd.DataFrame(columns=essential_cols + [specialty_col]) # Empty DF


# %% [markdown]
# ## 1. Define Scope and Filter Data
# Select target drugs (e.g., high-cost/volume) and optionally target specialties for outlier analysis. Filter the main DataFrame accordingly.

# %% [code]
# --- Define Targets ---
# Option 1: Use rankings from previous tickets if available
target_drugs_for_outliers = []
if 'generic_ranked_cost' in locals() and not generic_ranked_cost.empty:
    # Example: Choose top 5 generic drugs by cost
    target_drugs_for_outliers.extend(generic_ranked_cost['Gnrc_Name'].head(5).tolist())
if 'brand_ranked_cost' in locals() and not brand_ranked_cost.empty:
     # Example: Choose top 5 brand drugs by cost
     target_drugs_for_outliers.extend(brand_ranked_cost['Brnd_Name'].head(5).tolist())

# Option 2: Manually define list if needed
if not target_drugs_for_outliers:
    print("Ticket 2 drug rankings not found. Using example placeholder drugs for outlier detection.")
    print("--- PLEASE UPDATE 'target_drugs_for_outliers' with actual drug names ---")
    target_drugs_for_outliers = ['OZEMPIC', 'ELIQUIS', 'LISINOPRIL', 'AMLODIPINE BESYLATE'] # Example placeholders

# Ensure unique drug names
target_drugs_for_outliers = list(set(target_drugs_for_outliers))
print(f"Target drugs for outlier analysis: {target_drugs_for_outliers}")

# Define target specialties (optional)
# Example: ['Internal Medicine', 'Cardiology', 'Family Practice']
# Set to empty list or None to analyze across all specialties for the target drugs
target_specialties = []
if specialty_col_exists and target_specialties:
     print(f"Target specialties for analysis: {target_specialties}")
elif specialty_col_exists:
     print(f"Analyzing across all specialties for target drugs (no specific specialty filter applied).")
else:
     print("No specialty data available or no target specialties defined.")

# --- Filter Data ---
if not df_t5.empty:
    # Filter by drug name (check both generic and brand columns)
    df_filtered = df_t5[
        df_t5['Gnrc_Name'].isin(target_drugs_for_outliers) |
        df_t5['Brnd_Name'].isin(target_drugs_for_outliers)
    ].copy() # Use copy to avoid SettingWithCopyWarning

    # Filter by specialty if defined and available
    if specialty_col_exists and target_specialties:
        df_filtered = df_filtered[df_filtered[specialty_col].isin(target_specialties)]

    print(f"Filtered data for outlier analysis contains {df_filtered.shape[0]} rows.")

    if df_filtered.empty:
        print("Warning: No data found matching the specified drug/specialty filters.")
else:
    print("Skipping filtering as initial DataFrame is empty.")
    df_filtered = pd.DataFrame()

# %% [markdown]
# ## 2. Outlier Detection Method: Interquartile Range (IQR)
# We will use the IQR method to identify outliers. A data point is considered an outlier if it falls below $Q1 - 1.5 \times IQR$ or above $Q3 + 1.5 \times IQR$. This method is generally more robust to skewed distributions common in cost and volume data compared to Z-scores.

# %% [markdown]
# ## 3. Cost Per Claim (CPC) Outlier Detection
# Identify providers with significantly high `Cost_Per_Claim` for specific drugs (and optionally within specific specialties), compared to their peers. We'll filter out records with very few claims before calculating CPC outliers.

# %% [code]
# --- Configuration ---
min_claims_for_cpc_outlier = 5 # Minimum claims for a provider's drug record to be included in CPC outlier analysis

# --- Analysis ---
cpc_outliers_list = []

if not df_filtered.empty:
    print(f"\nDetecting Cost Per Claim outliers (using IQR, min claims threshold: {min_claims_for_cpc_outlier})...")

    # Define grouping columns based on whether specialty analysis is performed
    grouping_cols = ['Gnrc_Name', 'Brnd_Name']
    if specialty_col_exists and not target_specialties: # Group if col exists but no specific target specialties selected
        grouping_cols.append(specialty_col)
    elif specialty_col_exists and target_specialties: # Also group if specific specialties were targeted
        grouping_cols.append(specialty_col)

    # Group data
    grouped_for_cpc = df_filtered.groupby(grouping_cols)

    for name, group in grouped_for_cpc:
        # Filter group based on minimum claims threshold
        group_filtered = group[group['Tot_Clms'] >= min_claims_for_cpc_outlier]

        if len(group_filtered) < 5: # Need a few points to calculate IQR reasonably
            continue # Skip groups with too few data points after claim filtering

        # Calculate IQR for Cost_Per_Claim
        Q1 = group_filtered['Cost_Per_Claim'].quantile(0.25)
        Q3 = group_filtered['Cost_Per_Claim'].quantile(0.75)
        IQR = Q3 - Q1

        # Define outlier bounds (we are mostly interested in high CPC outliers)
        # lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Avoid issues with IQR=0 (all values the same)
        if IQR == 0 and Q1 == Q3 :
             # If all values are the same after filtering, only flag if > 0
             # Or consider a different logic if needed
             # For now, we only look for values above Q3 + 1.5*IQR, so IQR=0 means no outliers above Q3
             pass
        elif IQR == 0 and Q1 != Q3: # Should not happen with quantile logic but safety check
             continue


        # Identify outliers (only upper bound for CPC typically)
        # outliers_in_group = group_filtered[
        #     (group_filtered['Cost_Per_Claim'] < lower_bound) |
        #     (group_filtered['Cost_Per_Claim'] > upper_bound)
        # ]
        outliers_in_group = group_filtered[group_filtered['Cost_Per_Claim'] > upper_bound]


        if not outliers_in_group.empty:
            outlier_df = outliers_in_group[['Prscrbr_NPI', 'Gnrc_Name', 'Brnd_Name', specialty_col if specialty_col_exists else None, 'Tot_Clms', 'Cost_Per_Claim']].copy().dropna(axis=1, how='all') # Remove placeholder cols if not used
            outlier_df['Outlier_Type'] = 'High Cost Per Claim'
            outlier_df['Group_Q1'] = Q1
            outlier_df['Group_Q3'] = Q3
            outlier_df['Group_IQR'] = IQR
            outlier_df['Upper_Bound'] = upper_bound
            cpc_outliers_list.append(outlier_df)

    if cpc_outliers_list:
        cpc_outliers_df = pd.concat(cpc_outliers_list, ignore_index=True)
        print(f"\nFound {cpc_outliers_df.shape[0]} potential High Cost Per Claim outliers:")
        display(cpc_outliers_df.sort_values(by='Cost_Per_Claim', ascending=False).head(20))
    else:
        print("\nNo significant High Cost Per Claim outliers found based on the criteria.")
        cpc_outliers_df = pd.DataFrame()

else:
    print("Skipping Cost Per Claim outlier detection as filtered data is empty.")
    cpc_outliers_df = pd.DataFrame()


# %% [markdown]
# ## 4. Provider Volume Outlier Detection
# Identify providers with significantly high `Tot_Clms` for specific drugs (and optionally within specific specialties), compared to their peers prescribing the same drug.

# %% [code]
volume_outliers_list = []

if not df_filtered.empty:
    print("\nDetecting Provider Volume outliers (using IQR)...")

    # Define grouping columns
    grouping_cols_vol = ['Gnrc_Name', 'Brnd_Name']
    if specialty_col_exists and not target_specialties:
        grouping_cols_vol.append(specialty_col)
    elif specialty_col_exists and target_specialties:
        grouping_cols_vol.append(specialty_col)

    # Aggregate Total Claims per Provider within each group
    provider_volume = df_filtered.groupby(grouping_cols_vol + ['Prscrbr_NPI'])['Tot_Clms'].sum().reset_index()

    # Now group by drug (and specialty) to find outliers among providers
    grouped_for_vol = provider_volume.groupby([col for col in grouping_cols_vol if col != 'Prscrbr_NPI'])

    for name, group in grouped_for_vol:
        if len(group) < 5: # Need a few providers for comparison
            continue

        # Calculate IQR for Total Claims per provider in this group
        Q1 = group['Tot_Clms'].quantile(0.25)
        Q3 = group['Tot_Clms'].quantile(0.75)
        IQR = Q3 - Q1

        # Define outlier bounds (mostly interested in high volume)
        # lower_bound_vol = Q1 - 1.5 * IQR
        upper_bound_vol = Q3 + 1.5 * IQR

        if IQR == 0: # Avoid issues if IQR is 0
             continue

        # Identify outliers (only upper bound for volume)
        outliers_in_group = group[group['Tot_Clms'] > upper_bound_vol]

        if not outliers_in_group.empty:
            # Add context columns back if needed (Gnrc_Name, Brnd_Name, Specialty)
            outlier_df = outliers_in_group.copy()
            # If name is a tuple (multiple grouping cols), assign them back
            if isinstance(name, tuple):
                 for i, col_name in enumerate([col for col in grouping_cols_vol if col != 'Prscrbr_NPI']):
                      outlier_df[col_name] = name[i]
            else: # Single grouping col
                 outlier_df[[col for col in grouping_cols_vol if col != 'Prscrbr_NPI'][0]] = name

            outlier_df['Outlier_Type'] = 'High Volume'
            outlier_df['Group_Q1'] = Q1
            outlier_df['Group_Q3'] = Q3
            outlier_df['Group_IQR'] = IQR
            outlier_df['Upper_Bound'] = upper_bound_vol
            volume_outliers_list.append(outlier_df)


    if volume_outliers_list:
        volume_outliers_df = pd.concat(volume_outliers_list, ignore_index=True)
        # Reorder columns for clarity
        cols_order = ['Prscrbr_NPI'] + [col for col in grouping_cols_vol if col != 'Prscrbr_NPI'] + ['Tot_Clms', 'Outlier_Type', 'Group_Q1', 'Group_Q3', 'Group_IQR', 'Upper_Bound']
        volume_outliers_df = volume_outliers_df[cols_order]
        print(f"\nFound {volume_outliers_df.shape[0]} potential High Volume provider outliers:")
        display(volume_outliers_df.sort_values(by='Tot_Clms', ascending=False).head(20))
    else:
        print("\nNo significant High Volume provider outliers found based on the criteria.")
        volume_outliers_df = pd.DataFrame()

else:
    print("Skipping Provider Volume outlier detection as filtered data is empty.")
    volume_outliers_df = pd.DataFrame()


# %% [markdown]
# ## 5. Visualization
# Use box plots to visualize the distributions and outliers identified.

# %% [code]
if not df_filtered.empty:
    print("\nGenerating visualizations...")

    # --- Box Plot for Cost Per Claim ---
    # Use the data *before* filtering by min_claims for visualization to see full distribution
    print("\nCost Per Claim Distribution by Drug (subset of target drugs):")
    plt.figure(figsize=(15, 8))
    # Select a subset of drugs if too many, otherwise plot becomes unreadable
    plot_drugs_cpc = target_drugs_for_outliers[:10] # Max 10 drugs
    plot_data_cpc = df_filtered[
        df_filtered['Gnrc_Name'].isin(plot_drugs_cpc) |
        df_filtered['Brnd_Name'].isin(plot_drugs_cpc)
    ].copy()
    # Create a combined drug name column for plotting labels
    plot_data_cpc['Drug_Plot_Name'] = np.where(plot_data_cpc['Gnrc_Name'].isin(plot_drugs_cpc), plot_data_cpc['Gnrc_Name'], plot_data_cpc['Brnd_Name'])

    sns.boxplot(data=plot_data_cpc, y='Drug_Plot_Name', x='Cost_Per_Claim', orient='h', showfliers=True) # Show outliers
    plt.title(f'Distribution of Cost Per Claim for Target Drugs (Min Claims Filter NOT applied here)')
    plt.xlabel('Cost Per Claim ($)')
    plt.ylabel('Drug Name')
    # Consider log scale if data is highly skewed
    # plt.xscale('log')
    # plt.xlabel('Cost Per Claim ($) (Log Scale)')
    plt.tight_layout()
    plt.show()

    # --- Box Plot for Provider Volume per Drug ---
    if not volume_outliers_list: # Check if provider_volume was calculated
         print("\nSkipping Provider Volume visualization as no volume data was aggregated.")
    else:
        print("\nProvider Total Claims Distribution by Drug (subset of target drugs):")
        plt.figure(figsize=(15, 8))
        plot_drugs_vol = target_drugs_for_outliers[:10] # Max 10 drugs
        plot_data_vol = provider_volume[ # Use the aggregated provider volume data
             provider_volume['Gnrc_Name'].isin(plot_drugs_vol) |
             provider_volume['Brnd_Name'].isin(plot_drugs_vol)
        ].copy()
        plot_data_vol['Drug_Plot_Name'] = np.where(plot_data_vol['Gnrc_Name'].isin(plot_drugs_vol), plot_data_vol['Gnrc_Name'], plot_data_vol['Brnd_Name'])


        sns.boxplot(data=plot_data_vol, y='Drug_Plot_Name', x='Tot_Clms', orient='h', showfliers=True) # Show outliers
        plt.title('Distribution of Total Claims per Provider for Target Drugs')
        plt.xlabel('Total Claims per Provider (for the specific drug)')
        plt.ylabel('Drug Name')
        # Consider log scale if data is highly skewed
        # plt.xscale('log')
        # plt.xlabel('Total Claims per Provider (Log Scale)')
        plt.tight_layout()
        plt.show()

else:
    print("Skipping visualizations as filtered data is empty.")


# %% [markdown]
# ## 6. Summary & Next Steps
#
# - Defined target drugs (and optionally specialties) for outlier analysis.
# - Used the IQR method to identify potential outliers based on:
#     - High `Cost_Per_Claim` for individual drug records (filtering out low-volume records first).
#     - High total `Tot_Clms` per provider for specific drugs.
# - Generated lists of these potential outliers, including context (group stats, thresholds).
# - Visualized the distributions using box plots, highlighting the presence of outliers.
#
# **Next Steps:** The identified outliers represent providers or specific prescriptions that deviate significantly from the norm for a given drug (and potentially specialty). These are not necessarily errors but warrant further investigation. They could indicate specific patient populations, data entry issues, off-label use (for volume), prescribing of specific high-cost formulations (for CPC), or other factors. These outlier lists provide concrete points for targeted review or discussion during rebate negotiations (e.g., querying high CPC, understanding high volume drivers).

# %% [code]
print("\nTicket 5 Execution Complete.")
```

```python
# %% [markdown]
# # Ticket 7: Market Share & Usage Pattern Analysis
#
# **Description:** Define the specific drug(s) of interest ("our drug") and key competitors. Calculate market share based on Tot_Clms and potentially Tot_Drug_Cst within relevant therapeutic classes (if mapped) or compared to direct competitors identified in the data. Analyze usage patterns like average Tot_Day_Suply per claim for these drugs.
#
# **Goal:** Assess the competitive positioning and real-world usage patterns of key drugs, providing data points for market share-based rebate discussions.
#
# **Output:** Calculated market share percentages and usage pattern metrics (e.g., average days supply) in Jupyter.

# %% [markdown]
# ## Prerequisite Check
# Ensure the DataFrame `df` exists and contains necessary columns. Check if therapeutic class data (`df_merged_class`) is available if class-based analysis is desired.

# %% [code]
import pandas as pd
import numpy as np

# --- Check DataFrame and Columns ---
required_cols_t7 = ['Gnrc_Name', 'Brnd_Name', 'Tot_Clms', 'Tot_Drug_Cst', 'Tot_Day_Suply']
df_exists_t7 = 'df' in locals() or 'df' in globals()
class_data_exists = 'df_merged_class' in locals() or 'df_merged_class' in globals()
therapeutic_class_col = 'Therapeutic_Class' # Default name from Ticket 6

if df_exists_t7 and isinstance(df, pd.DataFrame):
    if all(col in df.columns for col in required_cols_t7):
        print("DataFrame 'df' found with required columns for Ticket 7.")
        df_t7 = df.copy() # Work on a copy
        # Ensure necessary columns are numeric, handling potential issues from source
        for col in ['Tot_Clms', 'Tot_Drug_Cst', 'Tot_Day_Suply']:
             df_t7[col] = pd.to_numeric(df_t7[col], errors='coerce')
        # Drop rows where key metrics needed for calculation are NaN
        df_t7.dropna(subset=['Tot_Clms', 'Tot_Drug_Cst', 'Tot_Day_Suply'], inplace=True)

        # Check for class data availability
        if class_data_exists and isinstance(df_merged_class, pd.DataFrame) and therapeutic_class_col in df_merged_class.columns:
            print(f"Therapeutic class data found in 'df_merged_class' (column: '{therapeutic_class_col}').")
            # Optional: Make a copy if using df_merged_class directly
            # df_t7_class = df_merged_class.copy()
        else:
            print("Therapeutic class data ('df_merged_class') not found or missing required column. Market share by class will not be available.")
            class_data_exists = False # Ensure flag is false

    else:
        missing_req_cols = [col for col in required_cols_t7 if col not in df.columns]
        print(f"Error: DataFrame 'df' is missing required columns: {missing_req_cols}")
        df_t7 = pd.DataFrame(columns=required_cols_t7) # Empty DF
elif not df_exists_t7 or not isinstance(df, pd.DataFrame):
    print("Error: DataFrame 'df' not found or is not a Pandas DataFrame.")
    print("Please ensure previous tickets have been executed successfully.")
    df_t7 = pd.DataFrame(columns=required_cols_t7) # Empty DF
else: # Should not be reached
     df_t7 = pd.DataFrame(columns=required_cols_t7) # Empty DF

# %% [markdown]
# ## 1. Define Drugs of Interest and Market Scope
# Specify "our drug(s)", competitor drug(s), and how the relevant market is defined (either the set of these drugs or a broader therapeutic class).

# %% [code]
# --- Define Drugs (USE UPPERCASE and ensure names match data) ---
# --- Replace these placeholders with your actual drug names ---
our_drugs = ['LENALIDOMIDE']
competitor_drugs = ['LENALIDOMIDE', 'POMALYST', 'THALIDOMIDE', 'NINLARO', 'POMALIDOMIDE', 'THALOMID', 'REVLIMID', 'SELINEXOR' ]

# Clean the lists (uppercase, strip whitespace)
our_drugs = [drug.upper().strip() for drug in our_drugs]
competitor_drugs = [drug.upper().strip() for drug in competitor_drugs]
all_defined_drugs = list(set(our_drugs + competitor_drugs))

print(f"Our Drug(s): {our_drugs}")
print(f"Competitor Drug(s): {competitor_drugs}")
print(f"All Defined Drugs for Competitor Set Analysis: {all_defined_drugs}")

# --- Define Market Scope ---
# Option 1: 'competitor_set' - Market is defined only by 'our_drugs' + 'competitor_drugs'
# Option 2: 'therapeutic_class' - Market is defined by a specific class (requires class data)
market_definition_type = 'competitor_set' # Choose 'competitor_set' or 'therapeutic_class'

# If using 'therapeutic_class', specify the class name(s)
# --- Replace with relevant class name(s) if using therapeutic_class ---
target_therapeutic_classes = ['GLP-1 RECEPTOR AGONISTS'] # Example

if market_definition_type == 'therapeutic_class':
     if not class_data_exists:
         print(f"Warning: Market definition set to 'therapeutic_class' but class data is not available. Reverting to 'competitor_set'.")
         market_definition_type = 'competitor_set'
     else:
         print(f"Market Scope: Therapeutic Class(es) - {target_therapeutic_classes}")
         # Clean target class names
         target_therapeutic_classes = [tc.upper().strip() for tc in target_therapeutic_classes]
elif market_definition_type == 'competitor_set':
     print(f"Market Scope: Competitor Set - Based on defined drugs")
else:
     print(f"Warning: Invalid 'market_definition_type'. Defaulting to 'competitor_set'.")
     market_definition_type = 'competitor_set'


# %% [markdown]
# ## 2. Filter Data for the Defined Market

# %% [code]
market_df = pd.DataFrame() # Initialize empty DataFrame

if not df_t7.empty:
    # Prepare drug name columns in df_t7 for matching
    df_t7['Gnrc_Name_Upper'] = df_t7['Gnrc_Name'].str.upper().str.strip()
    df_t7['Brnd_Name_Upper'] = df_t7['Brnd_Name'].str.upper().str.strip()

    if market_definition_type == 'competitor_set':
        print("Filtering data based on the defined competitor set...")
        market_df = df_t7[
            df_t7['Gnrc_Name_Upper'].isin(all_defined_drugs) |
            df_t7['Brnd_Name_Upper'].isin(all_defined_drugs)
        ].copy()
        # Create a 'Drug_Identifier' column for easier grouping later
        market_df['Drug_Identifier'] = market_df.apply(
            lambda row: row['Brnd_Name_Upper'] if row['Brnd_Name_Upper'] in all_defined_drugs else row['Gnrc_Name_Upper'],
            axis=1
        )
        # Ensure we only have drugs from our defined list
        market_df = market_df[market_df['Drug_Identifier'].isin(all_defined_drugs)]

    elif market_definition_type == 'therapeutic_class' and class_data_exists:
        print(f"Filtering data based on Therapeutic Class(es): {target_therapeutic_classes}...")
        # Assuming class data is in df_merged_class and needs similar prep
        if 'df_merged_class' not in locals() or df_merged_class is None:
             # Need to merge class info if not using df_merged_class directly
             # This requires the mapping file logic from Ticket 6 again - simplified here
             print("Merging therapeutic class info (requires mapping file)...")
             # Placeholder - Implement merging logic from Ticket 6 if needed
             # drug_class_map_df = pd.read_csv(...)
             # df_t7 = pd.merge(...)
             # ... check df_t7 for therapeutic_class_col ...
             print("Error: Class data merge not implemented here. Cannot proceed with class-based market share.")
             market_definition_type = 'competitor_set' # Fallback
             market_df = df_t7[ # Re-filter using competitor set
                    df_t7['Gnrc_Name_Upper'].isin(all_defined_drugs) |
                    df_t7['Brnd_Name_Upper'].isin(all_defined_drugs)
                ].copy()
             market_df['Drug_Identifier'] = market_df.apply(
                lambda row: row['Brnd_Name_Upper'] if row['Brnd_Name_Upper'] in all_defined_drugs else row['Gnrc_Name_Upper'],
                axis=1
             )
             market_df = market_df[market_df['Drug_Identifier'].isin(all_defined_drugs)]

        else: # Assumes df_merged_class exists and is ready
             df_merged_class[therapeutic_class_col] = df_merged_class[therapeutic_class_col].str.upper().str.strip()
             market_df = df_merged_class[
                 df_merged_class[therapeutic_class_col].isin(target_therapeutic_classes)
             ].copy()
             # Prepare drug name columns if needed for Drug_Identifier
             market_df['Gnrc_Name_Upper'] = market_df['Gnrc_Name'].str.upper().str.strip()
             market_df['Brnd_Name_Upper'] = market_df['Brnd_Name'].str.upper().str.strip()
             # Define Drug_Identifier based on available names in the class
             market_df['Drug_Identifier'] = np.where(market_df['Brnd_Name'].notna(), market_df['Brnd_Name_Upper'], market_df['Gnrc_Name_Upper'])


    # --- Cleanup temporary columns ---
    df_t7.drop(columns=['Gnrc_Name_Upper', 'Brnd_Name_Upper'], errors='ignore', inplace=True)
    if not market_df.empty:
        market_df.drop(columns=['Gnrc_Name_Upper', 'Brnd_Name_Upper'], errors='ignore', inplace=True)
        print(f"Filtered market data contains {market_df.shape[0]} rows and {market_df['Drug_Identifier'].nunique()} unique drugs.")
        if market_df.empty:
             print("Warning: No data found matching the market definition criteria.")
    else:
         print("Warning: Market DataFrame is empty after filtering.")


# %% [markdown]
# ## 3. Calculate Market Share
# Calculate Volume Share (based on Total Claims) and Value Share (based on Total Cost).

# %% [code]
market_share_results = {}

if not market_df.empty:
    print("\nCalculating Market Share...")

    # Calculate overall market totals (denominator)
    market_total_claims = market_df['Tot_Clms'].sum()
    market_total_cost = market_df['Tot_Drug_Cst'].sum()

    print(f"Total Market Claims: {market_total_claims:,.0f}")
    print(f"Total Market Cost: ${market_total_cost:,.2f}")

    if market_total_claims > 0 and market_total_cost > 0:
        # Calculate totals for each drug in the market
        drug_totals = market_df.groupby('Drug_Identifier').agg(
            Drug_Total_Claims=('Tot_Clms', 'sum'),
            Drug_Total_Cost=('Tot_Drug_Cst', 'sum')
        ).reset_index()

        # Calculate share for each drug
        drug_totals['Volume_Share (%)'] = (drug_totals['Drug_Total_Claims'] / market_total_claims) * 100
        drug_totals['Value_Share (%)'] = (drug_totals['Drug_Total_Cost'] / market_total_cost) * 100

        print("\nMarket Share per Drug within Defined Market:")
        display(drug_totals.sort_values(by='Value_Share (%)', ascending=False))

        # Specifically store results for "our drugs"
        our_drugs_share = drug_totals[drug_totals['Drug_Identifier'].isin(our_drugs)]
        if not our_drugs_share.empty:
             print("\nMarket Share for Our Drug(s):")
             display(our_drugs_share)
             market_share_results = our_drugs_share.set_index('Drug_Identifier').to_dict('index')
        else:
             print("\nCould not find specified 'our drug(s)' within the filtered market data.")

    else:
        print("Cannot calculate market share because total market claims or cost is zero.")

else:
    print("Skipping market share calculation as filtered market data is empty.")


# %% [markdown]
# ## 4. Analyze Usage Patterns (Average Days Supply)
# Calculate and compare the average days supply per claim for "our drug(s)" versus competitors.

# %% [code]
usage_pattern_results = {}

if not market_df.empty:
    print("\nAnalyzing Usage Patterns (Average Days Supply per Claim)...")

    # Calculate Weighted Average Days Supply = sum(Total Days Supply) / sum(Total Claims)
    avg_days_supply = market_df.groupby('Drug_Identifier').agg(
        Total_Days_Supply=('Tot_Day_Suply', 'sum'),
        Total_Claims=('Tot_Clms', 'sum') # Recalculate claims sum here for alignment
    ).reset_index()

    # Calculate the weighted average
    avg_days_supply['Avg_Days_Supply_per_Claim'] = np.divide(
        avg_days_supply['Total_Days_Supply'],
        avg_days_supply['Total_Claims']
    )
     # Handle potential NaN results from division by zero if any Tot_Clms were 0
    avg_days_supply['Avg_Days_Supply_per_Claim'].fillna(0, inplace=True)

    print("\nAverage Days Supply per Claim (Weighted):")
    display(avg_days_supply[['Drug_Identifier', 'Avg_Days_Supply_per_Claim']].sort_values(by='Avg_Days_Supply_per_Claim', ascending=False))

    # Store results for comparison
    our_drugs_usage = avg_days_supply[avg_days_supply['Drug_Identifier'].isin(our_drugs)]
    competitor_usage = avg_days_supply[avg_days_supply['Drug_Identifier'].isin(competitor_drugs)]

    usage_pattern_results['our_drugs'] = our_drugs_usage.set_index('Drug_Identifier')[['Avg_Days_Supply_per_Claim']].to_dict('index')
    usage_pattern_results['competitors'] = competitor_usage.set_index('Drug_Identifier')[['Avg_Days_Supply_per_Claim']].to_dict('index')

    # --- Optional: Simple Average ---
    # market_df['Days_Supply_Per_Claim'] = np.divide(market_df['Tot_Day_Suply'], market_df['Tot_Clms'])
    # avg_days_supply_simple = market_df.groupby('Drug_Identifier')['Days_Supply_Per_Claim'].mean().reset_index()
    # print("\nAverage Days Supply per Claim (Simple Mean of Row-Level Ratios):")
    # display(avg_days_supply_simple.sort_values(by='Days_Supply_Per_Claim', ascending=False))


else:
    print("Skipping usage pattern analysis as filtered market data is empty.")

# %% [markdown]
# ## 5. Summary & Next Steps
#
# - Defined drugs of interest ("our drug(s)") and key competitors.
# - Defined the relevant market scope (either the competitor set or a therapeutic class).
# - Filtered the dataset to isolate data for the defined market.
# - Calculated market share based on both volume (Total Claims) and value (Total Drug Cost) for all drugs within the market, highlighting the share for "our drug(s)".
# - Calculated the weighted average days supply per claim for drugs in the market, allowing comparison between "our drug(s)" and competitors.
#
# **Next Steps:** These market share figures provide a clear picture of competitive positioning within the chosen market definition. Differences in market share (Volume vs. Value) can indicate price positioning relative to competitors. The average days supply metric offers insight into typical usage patterns and prescription lengths. These data points are valuable inputs for strategic planning, performance tracking, and formulating arguments for market share-based rebate negotiations.

# %% [code]
# Display final stored results (optional)
# print("\n--- Final Stored Results ---")
# print("Market Share:", market_share_results)
# print("Usage Patterns (Avg Days Supply):", usage_pattern_results)

print("\nTicket 7 Execution Complete.")
```
